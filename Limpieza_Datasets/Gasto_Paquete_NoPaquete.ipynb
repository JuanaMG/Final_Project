```python
import pandas as pd
from datetime import datetime
import warnings
import numpy as np
```

```python
def renomcolumna(df):
    df.rename(columns={'Unnamed: 0':'procedencia',
                      'Gasto total (millones de euros)':'gasto_total',
                      'Gasto por persona (euros)':'gasto_persona',
                      'Gasto por persona y día (euros)':'gasto_persona_dia'},inplace=True)
    df.insert(1, "Mes/año", "")
    df.insert(2, "Isla", "")
```

```python
renomcolumna(paquete)
```

```python
paquete.procedencia = paquete.procedencia.str.strip()
```

```python
def fecha(df):
    count = 0
    cambioiloc = 0
    lista_general = []
    for i,row in df.iterrows():
        lista_general.append(df.iloc[cambioiloc]["procedencia"])
        count += 1    
        if count == 13:
            cambioiloc = (i+1)
            count=0
    df["Mes/año"] = lista_general   

    for i, row in df.iterrows():
        if row['procedencia'] in lista_general:
            df.drop(i, inplace=True)
```

```python
fecha(paquete)
```

```python
paquete.reset_index(inplace=True)
```

```python
paquete.drop(columns = 'index', inplace = True)
```

```python
def lugar(df):
    count = 0
    cambioiloc = 0
    lista_general = []
    for i,row in df.iterrows():
        lista_general.append(df.iloc[cambioiloc]["procedencia"])
        count += 1    
        if count == 4:
            cambioiloc = (i+1)
            count=0
    df["Isla"] = lista_general   

    for i, row in df.iterrows():
        if row['procedencia'] in lista_general:
            df.drop(i, inplace=True)
        if row['procedencia'] == 'TOTAL':
            df.drop(i, inplace=True)
```

```python
lugar(paquete)
```

```python
def month_year(df):
    df['year'], df['month'] = df["Mes/año"].str.split('M', 1).str
    df.drop(['Mes/año'], axis = 1, inplace= True)
```

```python
month_year(paquete)
```

```python
def cambiar (strnum):

    strnum = strnum.replace(".","")
    return strnum.replace(",",".")
```

```python
paquete['gasto_persona'] = paquete['gasto_persona'].apply(cambiar)
paquete['gasto_persona'] = pd.to_numeric(paquete['gasto_persona'], errors = 'coerce')
```

```python
paquete['gasto_persona_dia'] = paquete['gasto_persona_dia'].apply(cambiar)
paquete['gasto_persona_dia'] = pd.to_numeric(paquete['gasto_persona_dia'], errors = 'coerce')
```
