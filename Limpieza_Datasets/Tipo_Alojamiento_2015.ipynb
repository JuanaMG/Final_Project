```python
import pandas as pd
import numpy as np
```

```python
alojamiento_2015 = pd.read_csv('../Limpieza_Datasets/data/Tipo_alojamiento_isla_2015.csv', sep = ';')
```

```python
alojamiento_2015 = alojamiento_2015.rename(columns={'Unnamed: 0':'tipo_alojamiento',
                                                   'Eivissa-Formentera':'Ibiza_Formentera'})
alojamiento_2015.insert(1, "Mes/año", "")
alojamiento_2015.tipo_alojamiento = alojamiento_2015.tipo_alojamiento.str.strip()
```

```python
alojamiento_2015.head(20)
```

```python
count = 0
cambioiloc = 0
listafecha = []
for i,row in alojamiento_2015.iterrows():
    listafecha.append(alojamiento_2015.iloc[cambioiloc]["tipo_alojamiento"])
    count += 1    
    if count == 6:
        cambioiloc = (i+1)
        count=0
```

```python
alojamiento_2015["Mes/año"] = listafecha
```

```python
for i, row in alojamiento_2015.iterrows():
    if row['tipo_alojamiento'] in listafecha:
        alojamiento_2015.drop(i, inplace=True)
```

```python
alojamiento_2015['year'], alojamiento_2015['month'] = alojamiento_2015["Mes/año"].str.split('M', 1).str
alojamiento_2015 = alojamiento_2015.drop(['Mes/año'], axis = 1)
alojamiento_2015.head(15)
```

Para unir los dataFrame de "tipo de alojamiento", se van a realizar los siguientes cambios:
* En la columna 'Tipo de alojamiento' se deben unificar conceptos para poder hacer la unión:
    - 'Hoteles y similares', ' Alojamiento de alquiler' y 'Alojamiento en propiedad' van a ser igual.
    - 'Alojamiento de familiares o amigos' se va a considerar que es igual a 'Otro alojamiento no de mercado', se cambia el nombre en la columna. --> Unificado a 'Alojamiento de familiares o amigos'.
    - 'Otros' y 'Otro alojamiento de mercado' van a considerarse mismos datos. -> Unificado a 'Otro alojamiento de mercado'

```python
alojamiento_2015["tipo_alojamiento"].replace({"Otros": "Otro alojamiento de mercado"}, inplace=True)
```

```python
alojamiento_pmi = alojamiento_2015[['tipo_alojamiento','year', 'month', 'Mallorca']]
alojamiento_pmi.insert(1, "Isla", "Mallorca")
alojamiento_pmi = alojamiento_pmi.rename(columns={'Mallorca':'Num_establecimientos'})
```

```python
alojamiento_mah = alojamiento_2015[['tipo_alojamiento','year', 'month', 'Menorca']]
alojamiento_mah.insert(1, "Isla", "Menorca")
alojamiento_mah = alojamiento_mah.rename(columns={'Menorca':'Num_establecimientos'})
```

```python
alojamiento_ibz = alojamiento_2015[['tipo_alojamiento','year', 'month', 'Ibiza_Formentera']]
alojamiento_ibz.insert(1, "Isla", "Ibiza_Formentera")
alojamiento_ibz = alojamiento_ibz.rename(columns={'Ibiza_Formentera':'Num_establecimientos'})
```

```python
alojamiento_ok = pd.concat([alojamiento_pmi, alojamiento_mah, alojamiento_ibz])
```
