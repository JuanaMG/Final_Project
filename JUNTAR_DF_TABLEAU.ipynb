### Juntar df 'estancia media' con 'numero_turistas_origen'

```python
estancias
```

```python
num_turistas_origen
```

```python
def origen(isla,year, month, pais):
    origen = num_turistas_origen[(num_turistas_origen["Isla"]== isla) & (num_turistas_origen["year"]==year)& (num_turistas_origen["month"]==month)&(num_turistas_origen["origen_turistas"]==pais)]
    origen.reset_index(inplace=True)
    #try:
    #print(origen["Num_turistas"][0])
    return origen["Num_turistas"][0]
    #except:
        #return np.nan
```

```python
columnas = list(num_turistas_origen.iloc(1)[0].unique())
columnas
```

```python
dfjunto = pd.DataFrame(columns=list(estancias.columns))
```

```python
dfjunto = pd.concat([dfjunto,estancias])
```

```python
for i in columnas:
    dfjunto[f"num_turistas_{i}"] = dfjunto.apply(lambda x: origen(x["Isla"],x["year"], x["month"], i), axis=1) 
```

```python
dfjunto.head(2)
```

### Juntar df 'dfjunto' con 'gasto_persona_dia'

```python
def gasto(isla,year, month):
    a = gasto_persona_dia[(gasto_persona_dia["Isla"]== isla)&
                      (gasto_persona_dia["year"]==year) &
                      (gasto_persona_dia["month"]==month)]
    a.reset_index(inplace=True)

    return (a["Gasto_persona"][0],a["Gasto_persona_día"][0])
```

```python
dfjunto[f"Gasto_persona"] = dfjunto.apply(lambda x: gasto(x["Isla"],x["year"], x["month"])[0], axis=1)
```

```python
dfjunto[f"Gasto_persona_día"] = dfjunto.apply(lambda x: gasto(x["Isla"],x["year"], x["month"])[1], axis=1)
```

```python
dfjunto.head(2)
```

### Juntar df 'dfjunto' con 'gasto_origen'

```python
gasto_origen_turista.head(2)
```

```python
def gasto_origen(isla,year, month,pais):
    b = gasto_origen_turista[(gasto_origen_turista["year"]==year) &
                             (gasto_origen_turista["month"]==month)&
                             (gasto_origen_turista["origen_turista"]==pais)]
    b.reset_index(inplace=True)

    return (b["Gasto_persona"][0])
```

```python
columnas_origen = list(gasto_origen_turista.iloc(1)[0].unique())
columnas_origen
```

```python
for i in columnas_origen:
    dfjunto[f"Gasto_persona_{i}"] = dfjunto.apply(lambda x: gasto_origen(x["Isla"],x["year"], x["month"], i), axis=1)
```

```python
def tipo_establecim(isla,year, month, tipo):
    c = total_general_establecimientos[(total_general_establecimientos["Isla"]== isla) & 
                                 (total_general_establecimientos["year"]==year)& 
                                 (total_general_establecimientos["month"]==month) & 
                                 (total_general_establecimientos["tipo_alojamiento"]==tipo)]
    c.reset_index(inplace=True)
    return (c["Num_establecimientos"][0])
```

```python
columnas_tipo = list(total_general_establecimientos.iloc(1)[0].unique())
columnas_tipo
```

```python
for i in columnas_tipo:
    dfjunto[f"num_{i}"] = dfjunto.apply(lambda x: tipo_establecim(x["Isla"],x["year"], x["month"], i), axis=1) 
```
