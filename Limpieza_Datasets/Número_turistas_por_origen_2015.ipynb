```python
import pandas as pd
```

```python
num_turistas_2015 = pd.read_csv('../Final_Project/data/Número_turistas_y_origen_2015.csv', sep = ';')
```

```python
num_turistas_2015 = num_turistas_2015.rename(columns={'Unnamed: 0':'origen_turistas',
                                           'Eivissa-Formentera':'Ibiza_Formentera'})
```

```python
num_turistas_2015.insert(1, "Mes/año", "")
```

```python
count = 0
cambioiloc = 0
listafecha = []
for i,row in num_turistas_2015.iterrows():
    listafecha.append(num_turistas_2015.iloc[cambioiloc]["origen_turistas"])
    count += 1    
    if count == 14:
        cambioiloc = (i+1)
        count=0
```

```python
num_turistas_2015["Mes/año"] = listafecha
```

```python
for i, row in num_turistas_2015.iterrows():
    if row['origen_turistas'] in listafecha:
        num_turistas_2015.drop(i, inplace=True)
```

```python
num_turistas_2015 = num_turistas_2015[~num_turistas_2015.origen_turistas.str.contains('TOTAL' or 'EXTRANJERO')]
```

```python
num_turistas_2015['origen_turistas'] = num_turistas_2015['origen_turistas'].str.replace("\(\d+\)", "").str.replace("\d+", "").str.replace(".", "").str.strip()
```
