```python
def espacios (df):
    df.Isla.str.strip()
```

```python
def inicio(df):    
    df.Isla.str.strip()
    df.insert(1, "Mes/año", "")
```

```python
def fecha(df):
    count = 0
    cambioiloc = 0
    lista_general = []
    for i,row in df.iterrows():
        lista_general.append(df.iloc[cambioiloc]["Isla"])
        count += 1    
        if count == 4:
            cambioiloc = (i+1)
            count=0
    df["Mes/año"] = lista_general   

    for i, row in df.iterrows():
        if row['Isla'] in lista_general:
            df.drop(i, inplace=True)
```

```python
def month_year(df):
    df['year'], df['month'] = df["Mes/año"].str.split('M', 1).str
    df.drop(['Mes/año'], axis = 1, inplace= True)
```

```python
estancia = estancia.rename(columns={'España (Otras CCAA)':'España'})
```

```python
def replacee(s):
    i=str(s).find(',')
    if(i>0):
        return s[:i] + '.' + s[i+1:]
    else :
        return s
```

```python
estancia['España']=estancia['España'].apply(replacee)
```

```python
estancia['Extranjero']=estancia['Extranjero'].apply(replacee)
```

```python
estancia['España'] = pd.to_numeric(estancia['España'],errors = 'coerce')
```

```python
estancia['Extranjero'] = pd.to_numeric(estancia['Extranjero'],errors = 'coerce')
```

```python
def espacios (x):
    return x.strip()
```

```python
estancia.Isla = estancia.Isla.apply(espacios)
```

```python
estancia
```

```python
def sustituir(x):
    if x == 'Eivissa-Formentera': 
        return 'Ibiza_Formentera'
    else:
        return x
```

```python
estancia.Isla = estancia.Isla.apply(sustituir)
```
