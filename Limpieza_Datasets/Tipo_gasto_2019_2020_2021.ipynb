```python
import pandas as pd

import warnings
warnings.simplefilter('ignore')
```

```python
tipo_gasto_2021 = pd.read_csv('../Limpieza_Datasets/data/Gasto_partida_de_gasto_2019_2020_2021.csv', sep = ';')
tipo_gasto_2018 = pd.read_csv('../Limpieza_Datasets/data/Gasto_partida_de_gasto_2018_2017_2016_2015.csv', sep = ';')
```

```python
tipo_gasto_2021
```

```python
def renomcolumna(df):
    df.rename(columns={'Unnamed: 0':'procedencia'},inplace=True)
    df.insert(1, "Mes/año", "")
    df.insert(2, "Isla", "")
```

```python
renomcolumna(tipo_gasto_2021)
renomcolumna(tipo_gasto_2018)
```

```python
def fecha(df):
    count = 0
    cambioiloc = 0
    lista_general = []
    for i,row in df.iterrows():
        lista_general.append(df.iloc[cambioiloc]["procedencia"])
        count += 1    
        if count == 13:
            cambioiloc = (i+1)
            count=0
    df["Mes/año"] = lista_general
    
    for i, row in df.iterrows():
        if row['procedencia'] in lista_general:
            df.drop(i, inplace=True)
```

```python
fecha(tipo_gasto_2021)
```

```python
fecha(tipo_gasto_2018)
```

```python
def indice(df):
    df.reset_index(inplace=True)
    df.drop(columns = 'index', inplace = True)
```

```python
indice(tipo_gasto_2021)
```

```python
indice(tipo_gasto_2018)
```

```python
tipo_gasto_2021.procedencia = tipo_gasto_2021.procedencia.str.strip()
```

```python
tipo_gasto_2018.procedencia = tipo_gasto_2018.procedencia.str.strip()
```

```python
def lugar(df):
    count = 0
    cambioiloc = 0
    lista_general = []
    for i,row in df.iterrows():
        lista_general.append(df.iloc[cambioiloc]["procedencia"])
        count += 1    
        if count == 4:
            cambioiloc = (i+1)
            count=0
    df["Isla"] = lista_general   
    
    for i, row in df.iterrows():
        if row['procedencia'] in lista_general:
            df.drop(i, inplace=True)
        if row['procedencia'] == 'TOTAL':
            df.drop(i, inplace=True)
```

```python
lugar(tipo_gasto_2021)
```

```python
lugar(tipo_gasto_2018)
```

```python
def month_year(df):
    df['year'], df['month'] = df["Mes/año"].str.split('M', 1).str
    df.drop(['Mes/año'], axis = 1, inplace= True)
```

```python
month_year(tipo_gasto_2021)
```

```python
month_year(tipo_gasto_2018)
```

```python
tipo_gasto_2018
```

```python
tipo_gasto = pd.concat([tipo_gasto_2021, tipo_gasto_2018])
```

```python
tipo_gasto
```

```python
def sustituir(x):
    if x == 'Eivissa-Formentera': 
        return 'Ibiza_Formentera'
    else:
        return x
```

```python
tipo_gasto.Isla=tipo_gasto.Isla.apply(sustituir)
```

```python
tipo_gasto['procedencia'] = tipo_gasto['procedencia'].str.replace(r"\(.*\)","").str.rstrip()
```

```python
tipo_gasto = tipo_gasto.rename(columns={'GASTO TOTAL':'gasto',
                                        '1) Gasto incluido en paquete turístico':'borrar',
                                        '2) Gasto no incluido en paquete turístico':'eliminar',
                                        '2.1) Gasto en transporte origen/destino':'Transporte',
                                        '2.2) Gasto en alojamiento':'Alojamiento',
                                        '2.3) Gasto en manutención':'Manutención',
                                        '2.4) Gasto en actividades':'Actividades',
                                        '2.5) Resto de gastos':'Otros'})
```

```python
tipo_gasto
```

```python
tipo_gasto.drop(columns=['gasto', 'borrar', 'eliminar'], inplace = True)
```

```python
tipo_gasto
```

```python
def cambiar (strnum):

    strnum = strnum.replace(".","")
    return strnum.replace(",",".")
```

```python
tipo_gasto['Transporte'] = tipo_gasto['Transporte'].apply(cambiar)
tipo_gasto['Transporte'] = pd.to_numeric(tipo_gasto['Transporte'], errors = 'coerce')
```

```python
tipo_gasto['Alojamiento'] = tipo_gasto['Alojamiento'].apply(cambiar)
tipo_gasto['Alojamiento'] = pd.to_numeric(tipo_gasto['Alojamiento'], errors = 'coerce')
```

```python
tipo_gasto['Manutención'] = tipo_gasto['Manutención'].apply(cambiar)
tipo_gasto['Manutención'] = pd.to_numeric(tipo_gasto['Manutención'], errors = 'coerce')
```

```python
tipo_gasto['Actividades'] = tipo_gasto['Actividades'].apply(cambiar)
tipo_gasto['Actividades'] = pd.to_numeric(tipo_gasto['Actividades'], errors = 'coerce')
```

```python
tipo_gasto['Otros'] = tipo_gasto['Otros'].apply(cambiar)
tipo_gasto['Otros'] = pd.to_numeric(tipo_gasto['Otros'], errors = 'coerce')
```

```python
tipo_gasto['year'] = tipo_gasto['year'].astype(str).astype(int)
tipo_gasto['month'] = tipo_gasto['month'].astype(str).astype(int)
```
