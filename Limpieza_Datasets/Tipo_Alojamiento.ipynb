```python
import pandas as pd
```

```python
alojamiento = pd.read_csv('../Limpieza_Datasets/data/Tipo_alojamiento_isla.csv', sep = ';')
```

```python
alojamiento.head(15)
```

```python
alojamiento.drop(['ILLES BALEARS'], axis=1, inplace=True)
```

```python
alojamiento = alojamiento.rename(columns={'Unnamed: 0':'tipo_alojamiento',
                                                   'Eivissa-Formentera':'Ibiza_Formentera'})
alojamiento.insert(1, "Mes/año", "")
alojamiento.tipo_alojamiento = alojamiento.tipo_alojamiento.str.strip()
```

```python
count = 0
cambioiloc = 0
listafecha = []
for i,row in alojamiento.iterrows():
    listafecha.append(alojamiento.iloc[cambioiloc]["tipo_alojamiento"])
    count += 1    
    if count == 9:
        cambioiloc = (i+1)
        count=0
```

```python
alojamiento["Mes/año"] = listafecha
```

```python
alojamiento.head(10)
```

```python
for i, row in alojamiento.iterrows():
    if row['tipo_alojamiento'] in listafecha:
        alojamiento.drop(i, inplace=True)
```

```python
alojamiento = alojamiento[~alojamiento.tipo_alojamiento.str.contains('TOTAL')]
```

```python
alojamiento.head(15)
```

```python
alojamiento['year'], alojamiento['month'] = alojamiento["Mes/año"].str.split('M', 1).str
alojamiento = alojamiento.drop(['Mes/año'], axis = 1)
alojamiento.head(15)
```

```python
alojamiento.tipo_alojamiento.unique()
```

```python
alojamiento["tipo_alojamiento"].replace({'Otro alojamiento no de mercado': 'Alojamiento de familiares o amigos',}, inplace=True)
```

```python
#alojamiento = alojamiento.drop(alojamiento[alojamiento.tipo_alojamiento == '1. Alojamiento de mercado'].index)
```

```python
alojamiento = alojamiento.drop(alojamiento[(alojamiento.tipo_alojamiento == '1. Alojamiento de mercado')].index)
alojamiento = alojamiento.drop(alojamiento[(alojamiento.tipo_alojamiento == '2. Alojamiento de no mercado')].index)
```

```python
alojamiento
```

```python
alojamiento_pmi = alojamiento[['tipo_alojamiento','year', 'month', 'Mallorca']]
alojamiento_pmi.insert(1, "Isla", "Mallorca")
alojamiento_pmi = alojamiento_pmi.rename(columns={'Mallorca':'Num_establecimientos'})
```

```python
alojamiento_mah = alojamiento[['tipo_alojamiento','year', 'month', 'Menorca']]
alojamiento_mah.insert(1, "Isla", "Menorca")
alojamiento_mah = alojamiento_mah.rename(columns={'Menorca':'Num_establecimientos'})
```

```python
alojamiento_ibz = alojamiento[['tipo_alojamiento','year', 'month', 'Ibiza_Formentera']]
alojamiento_ibz.insert(1, "Isla", "Ibiza_Formentera")
alojamiento_ibz = alojamiento_ibz.rename(columns={'Ibiza_Formentera':'Num_establecimientos'})
```

```python
alojamiento_ibz.isna().sum()
```

```python
alojamiento_ok = pd.concat([alojamiento_pmi,alojamiento_mah,alojamiento_ibz])
```

```python
alojamiento_ok
```

```python
alojam_2015 = pd.read_csv('../Limpieza_Datasets/data/num_establecim_2015_yalimpio.csv')
```

```python
ok_df_bueno = pd.concat([alojamiento_ok,alojam_2015])
```

```python
ok_df_bueno
```

```python
ok_df_bueno.tipo_alojamiento.unique()
```

```python
ok_df_bueno["tipo_alojamiento"].replace({'Otro alojamiento no de mercado': 'Alojamiento de familiares o amigos',
                                        'Hoteles o similares':'Hoteles y similares',
                                        'Alojamiento en alquiler':'Alojamiento de alquiler',
                                        'Vivienda en propiedad':'Alojamiento en propiedad'}, inplace=True)
```

```python
ok_df_bueno.tipo_alojamiento.unique()
```

```python
ok_df_bueno.dtypes
```

```python
def cambiar (strnum):

    strnum = strnum.replace(".","")
    return strnum.replace(",",".")
```

```python
ok_df_bueno['Num_establecimientos'] = ok_df_bueno['Num_establecimientos'].apply(cambiar)
ok_df_bueno['Num_establecimientos'] = pd.to_numeric(ok_df_bueno['Num_establecimientos'], errors = 'coerce')
```

```python
ok_df_bueno['year'] = ok_df_bueno['year'].astype(str).astype(int)
ok_df_bueno['month'] = ok_df_bueno['month'].astype(str).astype(int)
```
